```python
ax=sns.countplot(x='Gender',data=df)
```

```python
ax = sns.countplot(x='Gender', data=df, palette='Set2')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_gender=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=sales_gender,palette='Set1')
```

```python
                                                                                          sales_gender = df.groupby(['Gender'], as_index=False)['Amount'].sum()
sales_gender = sales_gender.sort_values(by='Amount', ascending=False)

ax = sns.barplot(x='Gender', y='Amount', data=sales_gender, palette='Set1')

# ---- Add % labels on top of bars ----
total = sales_gender['Amount'].sum()

for p in ax.patches:
    height = p.get_height()
    percent = 100 * height / total
    ax.annotate(f'{percent:.1f}%',
                (p.get_x() + p.get_width() / 2, height),
                ha='center', va='bottom',
                fontsize=11, fontweight='bold')
```

```python
# Calculate counts and sort them
order = df['Age Group'].value_counts().sort_values(ascending=False).index

ax = sns.countplot(data=df, x='Age Group', palette='Set3', order=order)

# ---- Add % labels ----
total = len(df)

for p in ax.patches:
    count = p.get_height()
    percent = 100 * count / total

    ax.annotate(f'{percent:.1f}%',
                (p.get_x() + p.get_width() / 2, count),
                ha='center', va='bottom',
                fontsize=11, fontweight='bold')
```

```python
# Sort occupations in descending order
order = df['Occupation'].value_counts().sort_values(ascending=False).index

sns.set(rc={'figure.figsize': (20,5)})

# Apply a color palette
ax = sns.countplot(
    data=df,
    x='Occupation',
    order=order,
    palette='Set3'          # you can change: Set1, Set2, Dark2, Pastel1, etc.
)

# Add labels on bars
for bars in ax.containers:
    ax.bar_label(bars)

# Rotate labels for better readability (optional)
plt.xticks(rotation=45, ha='right')
```

```python
sales_occupation = df.groupby(['Occupation'], as_index=False)['Amount'].sum() \
                     .sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize': (20,5)})
ax = sns.barplot(data=sales_occupation, x='Occupation', y='Amount', palette='Set2')

# ---- Add % labels ----
total = sales_occupation['Amount'].sum()

for p in ax.patches:
    height = p.get_height()
    percent = (height / total) * 100

    ax.annotate(f'{percent:.1f}%',
                (p.get_x() + p.get_width() / 2, height),
                ha='center', va='bottom',
                fontsize=11, fontweight='bold')

# Optional: Rotate labels
plt.xticks(rotation=45, ha='right')
```

```python
sales_occupation = df.groupby(['Occupation'], as_index=False)['Amount'].sum() \
                     .sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize': (20,5)})
ax = sns.barplot(data=sales_occupation, x='Occupation', y='Amount', palette='Set2')

# ---- Add % labels ----
total = sales_occupation['Amount'].sum()

for p in ax.patches:
    height = p.get_height()
    percent = (height / total) * 100

    ax.annotate(f'{percent:.1f}%',
                (p.get_x() + p.get_width() / 2, height),
                ha='center', va='bottom',
                fontsize=11, fontweight='bold')

# Optional: Rotate labels
plt.xticks(rotation=45, ha='right')
```

```python
sns.set(rc={'figure.figsize':(20,5)})
ax=sns.countplot(data=df,x='Product_Category',palette='Set2')
for bars in ax.containers:
  ax.bar_label(bars)

plt.xticks(rotation=45,ha='right')
```

```python
# Create descending order based on counts
order = df['Product_Category'].value_counts().index

sns.set(rc={'figure.figsize': (20,5)})

ax = sns.countplot(
    data=df,
    x='Product_Category',
    order=order,
    palette='Set2'
)

# Add labels
for bars in ax.containers:
    ax.bar_label(bars)

# Optional: rotate labels for clarity
plt.xticks(rotation=45, ha='right')
```

```python
# Top 5 categories in descending order
top10 = df['Product_Category'].value_counts().head(10).index

sns.set(rc={'figure.figsize': (20,5)})

ax = sns.countplot(
    data=df,
    x='Product_Category',
    order=top10,
    palette='Set2'
)

# Add bar labels
for bars in ax.containers:
    ax.bar_label(bars)

# Rotate labels (optional)
plt.xticks(rotation=45, ha='right')
```

```python
# Top 10 product in descending order
top_product = df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sns.set(rc={'figure.figsize': (20,5)})
sns.barplot(data=top_product,x='Product_ID',y='Orders',palette='Set2')
# Rotate labels (optional)
plt.xticks(rotation=45, ha='right')
```
